```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import glob
plt.style.use("ggplot")
```

## 地点マスタを読み取る

```python
dfmaster=pd.read_csv("../rawdata/ame_master_20220701.csv", encoding="cp932")
dfmaster=dfmaster[dfmaster["都府県振興局"].isin(["愛知","岐阜","三重"])]
dfmaster
```

```python
dfmaster["lat"]=dfmaster["緯度(度)"]+ dfmaster["緯度(分)"]/60.0
dfmaster["lon"]=dfmaster["経度(度)"]+ dfmaster["経度(分)"]/60.0
```

```python
dfmaster_sel=dfmaster.set_index("観測所名")
dfmaster_sel
```

```python
directions=["北","北北東","北東","東北東","東","東南東","南東","南南東","南","南南西","南西","西南西","西","西北西","北西","北北西"]

winddirection_to_theata={
    direction:i*22.5 for i,direction in enumerate(directions)
}
winddirection_to_theata["静穏"]=np.nan

def winddirection_to_deg(x):
    if x is np.nan:
        return np.nan
    if x not in directions:
        return np.nan
    return winddirection_to_theata[x]
```

```python
station_names_dict={"恵那":"ena","稲武":"inabu","亀山":"kameyama","岡崎":"okazaki","大垣":"oogaki","新城":"shinshiro","多治見":"tajimi"}
```

```python
for i in station_names_dict:
    print(i)
```

## アメダス

```python
for jp_stname in station_names_dict:
    st_name=station_names_dict[jp_stname]
    files=sorted(glob.glob(f"../rawdata/amedas_obs_hourly_{st_name}_*_*.csv.gz"))

    pickup_cols=["precip","temp","wind","wind_direction"]

    list_df=[]
    for f in files:
        list_df.append(pd.read_csv(f, index_col=["date_time"], parse_dates=True)[pickup_cols])
    dfall=pd.concat(list_df)
    #数値列を実数型に変換
    for col in ["precip","temp","wind"]:
        dfall[col]=dfall[col].astype(str).str.replace("[-×\]\)\/]","").replace("",np.nan).astype("float")
    #風向きを変換
    dfall["wind_direction"]=dfall["wind_direction"].str.replace("[\)\]\ ]","")
    dfall["wind_direction"]=dfall["wind_direction"].apply(winddirection_to_deg)
    dfall["u_wind"]=dfall["wind"]*np.sin(np.deg2rad(dfall["wind_direction"]))
    dfall["v_wind"]=dfall["wind"]*np.cos(np.deg2rad(dfall["wind_direction"]))
    dfall["lat"]=dfmaster_sel.loc[jp_stname,"lat"]
    dfall["lon"]=dfmaster_sel.loc[jp_stname,"lon"]
    dfall.to_csv(f"../data/amedas_obs_hourly_{st_name}_2010_2020.csv.gz", compression="gzip")
    print(jp_stname,"end")
```

## 官署

```python
dfweather=pd.read_csv("../data/situ_obs_2010_2020.csv.gz",parse_dates=True, index_col="date_time")
```

```python
##気象データの加工
weather_cols=['Ps', 'P0', 'precip', 'temp', 'dew_temp','e_vapor', 'rh', 'wind','sunshine_hours','solar_radiation']
for col in weather_cols:
    dfweather[col]=dfweather[col].astype(str).str.replace("[-×\]\)]","")
    dfweather[col]=dfweather[col].replace("",np.nan)
    dfweather[col]=dfweather[col].astype("float64")
dfweather["solar_radiation"].fillna(0,inplace=True)
dfweather["precip"].fillna(0, inplace=True)
```

```python
dfweather["wind_direction"]=dfweather["wind_direction"].str.replace("[\ \]\)]","")
```

```python
directions=["北","北北東","北東","東北東","東","東南東","南東","南南東","南","南南西","南西","西南西","西","西北西","北西","北北西"]

winddirection_to_theata={
    direction:i*22.5 for i,direction in enumerate(directions)
}
winddirection_to_theata["静穏"]=np.nan

def replace(x):
    if x is np.nan:
        return np.nan
    if x=='×':
        return np.nan
    return winddirection_to_theata[x]

dfweather["wind_direction"]=dfweather["wind_direction"].apply(replace)
```

```python
dfweather["year"]=dfweather.index.year
dfweather["M"]=dfweather.index.month
dfweather["u_wind"]=dfweather["wind"]*np.sin(np.deg2rad(dfweather["wind_direction"]))
dfweather["v_wind"]=dfweather["wind"]*np.cos(np.deg2rad(dfweather["wind_direction"]))
```

```python
cols=["temp","precip","rh","wind"]
dfweather_d=dfweather.resample("D").agg({"temp":"mean","precip":"sum","wind":"mean","rh":"mean","u_wind":"mean", "v_wind":"mean"})
dfweather_d.columns=[f"{col}_d" for col in dfweather_d.columns]
dfweather_d["temp_max"]=dfweather["temp"].resample("D").max()
dfweather_d["temp_min"]=dfweather["temp"].resample("D").min()
```

```python
dfweather_d["is_rainy"]=dfweather_d["precip_d"].apply(lambda x: 1 if x>=1 else 0)
```

```python
dfweather_d_yesterday=dfweather_d.copy()
dfweather_d_yesterday.index=dfweather_d_yesterday.index+pd.offsets.Day(1)
```

```python
dfweather_d_yesterday.columns=[f"{col}_1D_before" for col in dfweather_d_yesterday.columns] 
```

```python
dfweather_feature=pd.merge(dfweather_d,dfweather_d_yesterday, how="left",left_index=True, right_index=True)
```

```python
dfweather_feature["yyyy-mm-dd"]=dfweather_feature.index.strftime("%Y-%m-%d")
```

```python
for iyy in range(2010,2021):
    dfweather_feature.loc[f"{iyy}", "cum_temp"]=dfweather_feature.loc[f"{iyy}","temp_max"].cumsum()
```

```python
dfweather["yyyy-mm-dd"]=dfweather.index.strftime("%Y-%m-%d")
dfweather["date_time"]=dfweather.index
```

```python
dfweather_feature2=pd.merge(dfweather, dfweather_feature, how="left", on="yyyy-mm-dd")
```

```python
dfweather_feature2.index=dfweather_feature2["date_time"]
```

```python
features=['hour', 'P0', 'precip', 'temp', 'dew_temp',
       'e_vapor', 'rh', 'wind', 'wind_direction', 
       'solar_radiation',  'M', 'u_wind', 'v_wind', 'cum_temp', 
       'temp_d', 'precip_d', 'wind_d', 'rh_d', 'u_wind_d',
       'v_wind_d', 'temp_max', 'temp_min', 'is_rainy', 'temp_d_1D_before',
       'precip_d_1D_before', 'wind_d_1D_before', 'rh_d_1D_before',
       'u_wind_d_1D_before', 'v_wind_d_1D_before', 'temp_max_1D_before',
       'temp_min_1D_before', 'is_rainy_1D_before']
```

```python
dfweather_feature2=dfweather_feature2[features]
```
