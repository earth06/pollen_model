```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
plt.style.use("ggplot")
```

```python
df=pd.read_csv("../input/input.csv.gz",parse_dates=True, index_col=[0])
```

```python
df.dropna(subset=["pollen"],inplace=True)
```

```python
fig,ax=plt.subplots(figsize=(9,9))
ax.scatter(df["cum_temp"],df["pollen"])
```

```python
df_d=df.resample("D").mean().dropna(subset=["pollen"])
```

```python
for iyy in range(2010,2015):
    fig,ax=plt.subplots(figsize=(16,8))
    ax2=ax.twinx()
    df_d.loc[f"{iyy}","cum_temp"].plot(ax=ax)
    df_d.loc[f"{iyy}","pollen"].plot(ax=ax2)
#ax.scatter(df["cum_temp"],df["pollen"])
```
